### Übung 2 - Billboard

```python
billboard =pd.read_csv('billboard.csv')
```

```python
billboard
```

Bringen Sie Billboard in eine saubere Datenstruktur

```python
id_vars = ['year', 'artist', 'track', 'time', 'date.entered']

billboard_tidy = billboard.melt(id_vars = id_vars, var_name= "week", value_name="rank")

billboard_tidy.head()
```

Nutzen Sie isnull() um zu bestimmen, wieivie tracks nur 1 Woche platziert waren
Anmerkung isnull() produziert als Ergebnis  eine series (...kein Dataframe); die erste Spalte ist der Index !

```python

# Schritt 1: Identifizieren der Wochen-Spalten
wk_cols = [col for col in billboard.columns if col.startswith('wk')]

# Schritt 2: Zählen der NaN-Werte pro Track
num_nulls_per_track = billboard[wk_cols].isnull().sum(axis=1)

# Schritt 3: Berechnen der Anzahl der Wochen, in denen ein Track platziert war
total_weeks = len(wk_cols)
num_weeks_per_track = total_weeks - num_nulls_per_track

# Schritt 4: Identifizieren der Tracks, die nur 1 Woche platziert waren
one_week_tracks = num_weeks_per_track == 1

# Schritt 5: Zählen der Tracks, die nur 1 Woche platziert waren
num_one_week_tracks = one_week_tracks.sum()

print(f"Anzahl der Tracks, die nur 1 Woche platziert waren: {num_one_week_tracks}")
```

**Umgang mit kategorialen Daten**

Umwandeln ordinale Daten

```python
df_kat = pd.DataFrame([['green', 'M', 10.1],
                   ['red', 'L', 13.5],
                   ['blue', 'XL', 15.3]])

df_kat.columns = ['color', 'size', 'price']
df_kat
```

```python
size_mapping = {'XL': 3,
                'L': 2,
                'M': 1}

df_kat['size'] = df_kat['size'].map(size_mapping)
df_kat
```

Umwandeln kategoriale Daten mit One Hot Encoding

```python
df_ohe = pd.get_dummies(df_kat, drop_first=True)
df_ohe
```
