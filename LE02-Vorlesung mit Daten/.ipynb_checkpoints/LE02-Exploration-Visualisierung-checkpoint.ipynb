**2 - Lagemaße**

Achtung bei Statistik Funktionen - 'NaN' stören die Berechnung von Lagemaßen und die Darstellung von Histogrammen

Der Datensatz ohne 'fare' wird gelöscht.

```python
df_titanic.dropna(subset=['fare'], inplace=True)
```

```python
numeric_features = ['fare']

for f in numeric_features:
    print('Lagemaße', f,': ', "min %0.3f mean %0.3f median %0.3f mode %0.3f max %0.3f" 
          %(df_titanic[f].min(), df_titanic[f].mean(), df_titanic[f].median(), 
            df_titanic[f].mode(), df_titanic[f].max()))
```

**3 - Verteilung**

Verteilung des Merkmals 'fare'

```python
hist, ax = plt.subplots()

ax = df_titanic['fare'].plot.hist(bins=30)
ax.set_ylabel('Häufigkeit')
plt.show()
```

Verteilung des Merkmals 'fare' für die zwei Klassen 'survived'/ 'not survived'

```python
hist, ax = plt.subplots()
ax.set_xlim(0,600)
ax.set_ylim(0.1,600)
ax = df_titanic['fare'].loc[df_titanic['survived']==1].plot.hist(bins=30)
ax.set_title('survived')
plt.show()
```

```python
hist, ax = plt.subplots()
ax.set_xlim(0,600)
ax.set_ylim(0.1,600)
ax = df_titanic['fare'].loc[df_titanic['survived']==0].plot.hist(bins=15)
ax.set_title('not survived')
plt.show()
```

**Ausreisser analysieren**

```python
hist, ax = plt.subplots()
ax.set_yscale("log")
ax.set_xlim(0,600)
ax.set_ylim(0.1,1000)
ax = df_titanic['fare'].loc[df_titanic['survived']==1].plot.hist(bins=30)
ax.set_title('survived')
plt.show()

#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_yscale.html #matplotlib.axes.Axes.set_yscale
```

```python

ax = df_titanic.boxplot(['fare'], by ='survived')
ax.set_title('')
ax.set_ylabel('fare')
plt.show()
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html
```
