```python
import pandas as pd

# Daten einlesen
df1 = pd.read_csv("ds_I.csv")
df2 = pd.read_csv("ds_II.csv")
df3 = pd.read_csv("ds_III.csv")
df4 = pd.read_csv("ds_IV.csv")

# DataFrames in eine Liste packen
dataframes = [df1, df2, df3, df4]

# Leere Liste für die Ergebnisse
results = []

# Berechnungen durchführen und die Ergebnisse speichern
for df in dataframes:
    mean_x = round(df['x'].mean(), 2)
    var_x = round(df['x'].var(), 2)
    mean_y = round(df['y'].mean(), 2)
    var_y = round(df['y'].var(), 2)
    
    # Ergebnis pro Datensatz
    results.append([mean_x, var_x, mean_y, var_y])

# DataFrame mit den berechneten Werten
df_results = pd.DataFrame(results, columns=["x Mittelwert", "x Varianz", "y Mittelwert", "y Varianz"], index=["I", "II", "III", "IV"])

# MultiIndex für Zeilen
df_results = df_results.T
df_results.index = pd.MultiIndex.from_product([['x', 'y'], ['Mittelwert', 'Varianz']])

# Ausgabe des DataFrames
print(df_results)
```
